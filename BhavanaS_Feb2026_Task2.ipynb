```python
# =====================================================
# Data Science Internship – February 2026
# Task 2: Logic Building Task – 2
# Name: Bhavana S
# =====================================================


# -----------------------------------------------------
# Problem 1: Unique Words in a Sentence
# -----------------------------------------------------
print("\nProblem 1: Unique Words in a Sentence")

sentence = "python is easy and python is powerful"

# Split sentence into words
words = sentence.split()

# Convert list to set to remove duplicates
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)


# -----------------------------------------------------
# Problem 2: Highest Salary from Employee Data
# -----------------------------------------------------
print("\nProblem 2: Highest Salary from Employee Data")

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(employees, key=employees.get)

print("Highest Salary:", highest_employee, "-", employees[highest_employee])


# -----------------------------------------------------
# Problem 3: Find Maximum and Minimum Values
# -----------------------------------------------------
print("\nProblem 3: Find Maximum and Minimum Values")

numbers = [45, 22, 89, 10, 66]

# Initialize max and min
maximum = numbers[0]
minimum = numbers[0]

# Compare each element
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)


# -----------------------------------------------------
# Problem 4: Count Products Above a Price Threshold
# -----------------------------------------------------
print("\nProblem 4: Count Products Above 1000")

prices = [450, 1200, 899, 1500, 300]

count = 0
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)


# -----------------------------------------------------
# Problem 5: Calculate Attendance Percentage
# -----------------------------------------------------
print("\nProblem 5: Calculate Attendance Percentage")

attendance = ["P", "P", "A", "P", "P"]

# Count present days
present_days = attendance.count("P")
total_days = len(attendance)

attendance_percentage = (present_days / total_days) * 100

print("Attendance Percentage:", attendance_percentage)


# -----------------------------------------------------
# Problem 6: Remove Duplicate Phone Numbers
# -----------------------------------------------------
print("\nProblem 6: Remove Duplicate Phone Numbers")

phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to set to remove duplicates
unique_phone_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_phone_numbers)


# -----------------------------------------------------
# Problem 7: Count Character Frequency
# -----------------------------------------------------
print("\nProblem 7: Count Character Frequency")

text = "python"
char_frequency = {}

for char in text:
    if char in char_frequency:
        char_frequency[char] += 1
    else:
        char_frequency[char] = 1

print(char_frequency)


# -----------------------------------------------------
# Problem 8: Convert List to Tuple
# -----------------------------------------------------
print("\nProblem 8: Convert List to Tuple")

my_list = [10, 20, 30]
my_tuple = tuple(my_list)

print("Tuple:", my_tuple)


# -----------------------------------------------------
# Problem 9: Check if a Key Exists in a Dictionary
# -----------------------------------------------------
print("\nProblem 9: Check if a Key Exists in a Dictionary")

employee_data = {
    "Ravi": 75000,
    "Anita": 68000
}

if "Ravi" in employee_data:
    print("Employee exists")
else:
    print("Employee does not exist")


# -----------------------------------------------------
# Problem 10: Calculate Average Marks
# -----------------------------------------------------
print("\nProblem 10: Calculate Average Marks")

marks = [70, 75, 80, 65]

average_marks = sum(marks) / len(marks)

print("Average Marks:", average_marks)
```
